```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
import astropy.coordinates as coord
import astropy.units as u

# Sample data
table = """
Date,RA,DEC
March13,02:44:30.2,15:00:00.4
March16,02:46:47.7,15:10:58.1
March19,02:49:08.6,15:22:01.6
March22,02:49:08,15:29:28.3
March25,02:53:10.5,15:40:40.3
"""
tbl = ascii.read(table)

# Convert RA and DEC to angles
ra = coord.Angle(tbl["RA"], unit=u.hourangle).degree
ra = ra - 180  # Convert from 0-360 to -180-180
dec = coord.Angle(tbl["DEC"], unit=u.degree).degree

# Create figure and subplots
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the scatter plot
scatter = ax.scatter(np.radians(ra), np.radians(dec), s=50)  # Convert angles to radians for plotting
ax.grid(True)
ax.set_title('Map(Zoomed In)')
ax.set_xlabel('RA (radian)')
ax.set_ylabel('DEC (radian)')

# Annotate each point with its date
for i, txt in enumerate(tbl['Date']):
    ax.annotate(txt, (np.radians(ra[i]), np.radians(dec[i])), textcoords="offset points", xytext=(0,5), ha='center')

# Show the plot
plt.tight_layout()
plt.show()
```
